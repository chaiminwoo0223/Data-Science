# Import

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 수치형 데이터 시각화
## 1.히스토그램
- 수치형 데이터의 구간별 빈도수를 나타내는 그래프

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.histplot(data = titanic, x = 'age')
```

```python
# bins : 구간의 개수를 지정
sns.histplot(data = titanic, x = 'age', bins = 10)
```

```python
# hue : 범주별로 표현
# 회색 : 사망자수와 생존자수가 겹친 부분
sns.histplot(data = titanic, x = 'age', hue = 'alive')
```

```python
# multiple = 'stack' : 범주별로 누적해서 표현
sns.histplot(data = titanic, x = 'age', hue = 'alive', multiple = 'stack')
```

## 2.커널밀도추정함수 그래프
- 히스토그램을 매끄럽게 곡선으로 연결한 그래프

```python
sns.kdeplot(data = titanic, x = 'age', hue = 'alive', multiple = 'stack')
```

## 3.분포도
- 수치형 데이터 하나의 분포를 나타내는 그래프
- 파라미터만 조정하면, histplot()과 kdeplot()이 제공하는 기본 그래프를 모두 그릴 수 있다.

```python
# 히스토그램
sns.displot(data = titanic, x = 'age')
```

```python
# kind = 'kde' : 커널밀도추정함수
sns.displot(data = titanic, x = 'age', kind = 'kde')
```

```python
# kde = True : 히스토그램과 커널밀도함수를 동시에 그린다.
sns.displot(data = titanic, x = 'age', kde = True)
```

## 4.러그플롯
- 주변분포(marginal distribution)을 나타내는 그래프
- 단독으로 사용하기 보다는, 주로 다른 분포도 그래프와 함께 사용한다.

```python
sns.kdeplot(data = titanic, x = 'age')
sns.rugplot(data = titanic, x = 'age')
```

# 범주형 데이터 시각화
## 1.막대그래프
- 범주형 데이터 값에 따라, 수치형 데이터 값이 어떻게 달라지는지 파악할 때 사용한다.
- 평균 : 막대 높이, 신뢰구간 : 오차 막대

```python
# 신뢰구간이 넓다. = 표준편차가 크다.
sns.barplot(x = 'class', y = 'fare', data = titanic)
```

## 2.포인트플롯
- 범주형 데이터 값에 따라, 수치형 데이터 값이 어떻게 달라지는지 파악할 때 사용한다.
- 점과 선

```python
sns.pointplot(x = 'class', y = 'fare', data = titanic)
```

## 3.박스플롯
- 5가지 요약 수치 : 최솟값, 제1사분위 수(Q1), 제2사분위 수(Q2), 제3사분위 수(Q3), 최댓값
- 사분위 범위 수(IQR) : Q3 - Q1
- 이상치 : 최솟값보다 작은 값과 최댓값보다 큰 값

```python
sns.boxplot(x = 'class', y = 'age', data = titanic)
```

## 4.바이올린플롯
- 박스플롯 + 커널밀도추정함수

```python
sns.violinplot(x = 'class', y = 'age', data = titanic)
```

```python
# '성별'에 따른 '등급'별 '나이' 분포
sns.violinplot(x = 'class', y = 'age', data = titanic, split = True, hue = 'sex')
```

## 5.카운트플롯
- 범주형 데이터의 개수를 확인할 때 사용하는 그래프

```python
# X축
sns.countplot(x = 'class', data = titanic)
```

```python
# Y축
sns.countplot(y = 'class', data = titanic)
```

## 6.파이그래프
- 범주형 데이터별 비율을 알아볼 때 사용하기 좋은 그래프

```python
x = [10, 60, 30]         # 비율
labels = ['A', 'B', 'C'] # 레이블

plt.pie(x = x, labels = labels, autopct = '%.1f%%')
```

# 데이터 관계 시각화
## 1.히트맵
- 데이터 간 관계를 색상으로 표현한 그래프
- 비교해야 될 데이터가 많을 때 주로 사용한다.

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_pivot = flights.pivot(index = 'month', columns = 'year', values = 'passengers')
flights_pivot
```

```python
sns.heatmap(data = flights_pivot)
```

## 2.라인플롯
- 두 수치형 데이터 사이의 관계를 나타날 때 사용한다.
- X 파라미터에 전달한 값에 따라, Y 파라미터에 전달한 값의 평균과 95% 신뢰구간을 나타낸다.

```python
sns.lineplot(x = 'year', y = 'passengers', data = flights)
```

## 3.산점도
- 두 데이터의 관계를 점으로 표현하는 그래프



```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'time')
```

## 4.회귀선을 포함한 산점도 그래프
- 산점도 + 선형 회귀선

```python
# 신뢰구간 = 95%
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# 신뢰구간 = 99%
sns.regplot(x = 'total_bill', y = 'tip', data = tips, ci = 99)
```
